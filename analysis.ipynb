# Анализ результатов

```python
import matplotlib.pyplot as plt
```

После применения преобразования Хафа к исходным картинкам только для 4 изображений из 10 был осуществлён поворот на правильный угол. Это, скорее всего, связано с тем, что на картинках помимо рамок с текстом присутствуют другие объекты (например, фотографии лиц), в результате чего вычисляемые суммы для прямых становятся сильно зашумлёнными, а высокие значения дисперсий получаются и для строк с неправильным наклоном.

Чтобы другие объекты не оказывали такого сильного влияния, перед применением преобразования Хафа я использовал Canny edge detector, обсуждаемый на предыдущей лекции, из библиотеки OpenCV. В этом случае от таких объектов, как лица, на входе у алгоритма остаются только контуры, которые уже не оказывают такого большого влияния на результат преобразования. В итоге после применения комбинации этих двух преобразований все 10 изображений были повёрнуты правильно.

Посмотрим теперь на фактическую асимптотику и размер требуемой памяти алгоритма:

```python
for filename in ['time_from_area.png', 'time_from_N.png', 'memory.png']:
    plt.figure(figsize=(12, 8))
    plot = plt.imread(filename)
    plt.imshow(plot)
    plt.axis(False)
    plt.show()
```

Таким образом, размер требуемой памяти квадратично зависит от размера изображения, а его фактическая асимптотика $O(\log{N})$, где N - длина/ширина изображения.
